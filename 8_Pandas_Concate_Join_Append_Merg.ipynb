# append()

```python
df1 =pd.DataFrame({'name':['A','B','C','C'],'id':[101,102,103,101]})
df1
```

```python
df2 = pd.DataFrame({'name':['X','Y','Z','F'],'id':[201,202,203,201]})
df2
```

```python
#Append rows of `other` to the end of caller, returning a new object.
#syntax: first_df.append(sec_df)
df1.append(df2)
```

```python
df1.append(df2,ignore_index=True)
```

```python
df3 = pd.DataFrame({'name':['X','Y','Z','F'],'id':[201,202,203,201],
                   'age':[23,45,77,66]})
df3
```

```python
#if column is missing in any of df then it will give NaN
df1.append(df3)
```

```python
#we can append more than 1 df
df1.append([df2,df3,df2])
```

```python
df1.append(df1)
```

how : {'left', 'right', 'outer', 'inner', 'cross'}, default 'inner'
    Type of merge to be performed.

    * left: use only keys from left frame, similar to a SQL left outer join;
      preserve key order.
    * right: use only keys from right frame, similar to a SQL right outer join;
      preserve key order.
    * outer: use union of keys from both frames, similar to a SQL full outer
      join; sort keys lexicographically.
    * inner: use intersection of keys from both frames, similar to a SQL inner
      join; preserve the order of the left keys.
    * cross: creates the cartesian product from both frames, preserves the order
      of the left keys.
"""
